```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('Salary_Data.csv')
print(data)
```

```python
# X = data.iloc[:, -1].values
# Y = data.iloc[:, 0].values
```

```python
# x = np.arange(1, len(X))
# x = np.arange(1,4*np.pi,0.1)
x = np.arange(1,4*np.pi-2,0.5)
y = np.sin(x)
y2 = np.cos(x)

#sinewave using plot
f=plt.figure(figsize=(20, 6))

# salary plot adjust
f.set_figwidth(8)
f.set_figheight(3)

plt.plot(x, y, color = 'blue')
plt.xlabel('Radians')
plt.ylabel('Salary')
plt.title('Salary Sine-Wave')
plt.legend(['sin(x)'])


plt.show()
```

```python
#Cosinewave using plot

f=plt.figure(figsize=(20, 6))

# salary plot adjust
f.set_figwidth(8)
f.set_figheight(3)

plt.plot(x, y2, color = 'orange')
plt.xlabel('Radians')
plt.ylabel('Salary')
plt.title('Salary Cosine-Wave')
plt.legend(['cos(x)'])

plt.show()
```

```python
#Sinewave and Cosinewave using plot
f=plt.figure(figsize=(20, 6))

# salary plot adjust
f.set_figwidth(8)
f.set_figheight(3)

plt.plot(x, y, x, y2)
plt.xlabel('Radians')
plt.ylabel('Salary')
plt.title('Salary Sine-Wave & Cosine-Wave')
plt.legend(['sin(x)', 'cos(x)'], loc=3)


plt.show()
```

```python
#Sinewave and Cosinewave using subplot

plt.subplot(2, 1, 1)
plt.plot(x,y, label = "sin(x)", color = 'blue')
plt.title('Sinewave')
plt.legend(loc = 1)

plt.subplot(2, 1, 2)
plt.plot(x,y2, label = "cos(x)" , color = 'orange')
plt.title ('Cosinewave')
plt.legend (loc = 1)

plt.subplots_adjust(hspace=0.5)
plt.show()
```
